```python
# Import dependencies
import pandas as pd
import numpy as np
from datetime import datetime as dt
pd.set_option('max_colwidth', 400)
```

```python
#Loading the CSV files into df
employees_df = pd.read_csv(r"C:\Users\judy6\Downloads\challenge9\Starter_Code\data\employees.csv")
salaries_df = pd.read_csv(r"C:\Users\judy6\Downloads\challenge9\Starter_Code\data\salaries.csv")
dept_manager_df = pd.read_csv(r"C:\Users\judy6\Downloads\challenge9\Starter_Code\data\dept_manager.csv")
departments_df = pd.read_csv(r"C:\Users\judy6\Downloads\challenge9\Starter_Code\data\departments.csv")
titles_df = pd.read_csv(r"C:\Users\judy6\Downloads\challenge9\Starter_Code\data\titles.csv")
dept_emp = pd.read_csv(r"C:\Users\judy6\Downloads\challenge9\Starter_Code\data\dept_emp.csv")
```

```python
#combining the csvs
employee_salaries_df = employees_df.merge(salaries_df, on="emp_no")
```

```python
df = employee_salaries_df[["emp_no", "last_name", "first_name", "sex", "salary"]]
```

```python
#List the employee number, last name, first name, sex, and salary of each employee.
df
```

```python
#List the first name, last name, and hire date for the employees who were hired in 1986.
hired_in_1986 = employees_df[employees_df['hire_date'].str.contains('1986')]
df_2 = hired_in_1986[['first_name', 'last_name', 'hire_date']]
df_2
```

```python
#List the manager of each department along with their department number, department name, employee number, last name, and first name.
Manager_df = dept_manager_df.merge(departments_df, on="dept_no").merge(employees_df, left_on="emp_no", right_on="emp_no")
df_3 = Manager_df[["dept_no", "dept_name", "emp_no", "last_name", "first_name"]]
df_3
```

```python
#List the department number for each employee along with that employee’s employee number, last name, first name, and department name.
depart_df = dept_emp.merge(departments_df, on="dept_no").merge(employees_df, left_on="emp_no", right_on="emp_no")
df_4 =  depart_df[["dept_no", "emp_no", "last_name", "first_name", "dept_name"]]
df_4
```

```python
#List first name, last name, and sex of each employee whose first name is Hercules and whose last name begins with the letter B.
names = employees_df[(employees_df['first_name'] == 'Hercules') & (employees_df['last_name'].str.startswith('B'))]
df_5 = names[['first_name', 'last_name', 'sex']]
df_5
```

```python
#List each employee in the Sales department, including their employee number, last name, and first name.
emp = dept_emp.merge(departments_df, on="dept_no").merge(employees_df, left_on="emp_no", right_on="emp_no")
sales =  emp[emp["dept_name"] == "Sales"]
df_6 = sales[["emp_no", "last_name", "first_name"]]
df_6
```

```python
#List each employee in the Sales and Development departments, including their employee number, last name, first name, and department name
dev = dept_emp.merge(departments_df, on="dept_no").merge(employees_df, left_on="emp_no", right_on="emp_no")
sales_and_dev = dev[dev["dept_name"].isin(["Sales", "Development"])]
df_7 = sales_and_dev[["emp_no", "last_name", "first_name", "dept_name"]]
df_7
```

```python
#List the frequency counts, in descending order, of all the employee last names (that is, how many employees share each last name).
names = employees_df['last_name'].value_counts().reset_index()
names.columns = ['last_name', 'count']
df_8 = names.sort_values(by='count', ascending=False)
df_8
```
